# 결측값(Null) 알아보기

NaN 값에 대하여
- Null과 동일
- Pandas 에서는 Not a Number

```python
import pandas as pd
myfile = 'kpop_idols.csv'
df = pd.read_csv(myfile)
```

```python
df
```

```python
df.info()
```

df.info()로 대략 알아볼 수 있는데, <br>
위에는 1310 entries가 있으나, Non-Null Count가 그 수와 맞지 않는 만큼 NaN이 있다고 생각할 수 있음.

## 결측 값 다루기
Boolean 인덱싱 에서 True이면 NaN

```python
df.isna()
```

```python
df['Birthplace'].isnull()
```

```python
df['Birthplace'][df['Birthplace'].isnull()]
```

```python
df['Birthplace'][df['Birthplace'].isna()]
```

## NaN이 아닌 값 인덱싱

```python
df['Birthplace'].notnull()
```

```python
df['Birthplace'][df['Birthplace'].notnull()]
```

## Loc를 이용한 인덱싱

```python
df.loc[df['Birthplace'].notnull()]
```

```python
df.loc[df['Birthplace'].isnull(), ['Full Name', 'Date of Birth']]
```
