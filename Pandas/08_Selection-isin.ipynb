```python
import pandas as pd
myfile = 'kpop_idols.csv'
df = pd.read_csv(myfile)
```

```python
df[ df['Gender'] == 'F' ]
```

### loc를 이용하여 조건에 맞는 행 추출

```python
df.loc[df['Gender'] == 'F', ['Korean Name', 'Date of Birth']]
```

## isin 을 이용한 행 추출

### isin과 loc의 공부가 필요해 보임!!!

```python
my_condition = ['f(x)']
df['Group'].isin(my_condition)
```

```python
df.loc[df['Group'].isin(my_condition)]
```

```python
df.loc[df['Group'].isin(my_condition), ["Korean Name", "Country"]]
```
